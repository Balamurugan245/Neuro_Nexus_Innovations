```python
import pandas as pd
df=pd.read_csv("creditcard.csv")
df
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df.drop('Class',axis=1))
scaled_features=scaler.transform(df.drop('Class',axis=1))
new_df=pd.DataFrame(scaled_features,columns=df.columns[:-1])
new_df
```

```python
df["Class"].value_counts()
```

```python
x=new_df
y=df["Class"]
```

```python
#Balancing the dataset by oversampling
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(x, y)

y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
score = regressor.score(x_test, y_test)
score
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train, y_train)
y_pred = rf_regressor.score(x_test,y_test)
y_pred
```

```python
import numpy as np
y_pred=rf_regressor.predict(x_test).reshape(-1)
y_pred=np.round(y_pred)
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

```python
y=rf_regressor.predict(x_test).round()
import seaborn as sns
%matplotlib inline
from sklearn.metrics import confusion_matrix
result=confusion_matrix(y,y_test)
sns.heatmap(result,annot=True)
```

```python
#balancing data by undersampling
count_class_0, count_class_1 = df.Class.value_counts()
df_class_0 = df[df['Class'] == 0]
df_class_1 = df[df['Class'] == 1]
```

```python
df_class_0_under = df_class_0.sample(count_class_1)

df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)
df_test_under.Class.value_counts()
```

```python
x = df_test_under.drop(['Class'], axis=1)
y = df_test_under['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=15, stratify=y)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
score = regressor.score(x_test, y_test)
score
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train, y_train)
y_pred = rf_regressor.score(x_test,y_test)
y_pred
```

```python
#RandomForestRegressor gave the best score
import numpy as np
y_pred=rf_regressor.predict(x_test).reshape(-1)
y_pred=np.round(y_pred)
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

```python
y=rf_regressor.predict(x_test).round()
import seaborn as sns
%matplotlib inline
from sklearn.metrics import confusion_matrix
result=confusion_matrix(y,y_test)
sns.heatmap(result,annot=True)
```
